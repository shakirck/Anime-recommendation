**Data Cleaning**

```python
df = pd.DataFrame(anime)
df
```

```python
df.info
```

```python
df.dropna(subset=['rating', 'genre', 'members'], inplace=True)
```

```python
df
```

```python
df_cleaned = df.drop(columns=['anime_id', 'episodes'])
df_cleaned
```

**featere ebgineering**

```python
# Create a new column with only the primary genre from the list
df_cleaned['primary_genre'] = df_cleaned['genre'].apply(lambda x: x.split(',')[0])

# Create the target variable: 'recommendable'
# 1 = recommendable (rating >= 7.5), 0 = not
df_cleaned['recommendable'] = df_cleaned['rating'].apply(lambda x: 1 if x >= 7.5 else 0)

# Display the result to see the two new columns
print("DataFrame after adding 'primary_genre' and 'recommendable':")
print(df_cleaned.head().to_string())
```

**Prepare Data for Modeling**

```python
# Select the features the model will use: genre and popularity
model_df = df_cleaned[['primary_genre', 'members', 'recommendable']]

# Define features (X) and target (y)
X = model_df.drop('recommendable', axis=1)
y = model_df['recommendable']
```

```python
# Convert the text-based 'primary_genre' column into numerical columns
X_encoded = pd.get_dummies(X, columns=['primary_genre'])
print("--- Encoded Features Head (with members) ---")
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
print(f"\nData split into training ({X_train.shape[0]} rows) and testing ({X_test.shape[0]} rows) sets.")
```

**Train Model and Get Recommendations**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize and train the Decision Tree model
print("Training the Decision Tree model...")
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
accuracy = accuracy_score(y_test, dt_model.predict(X_test))
print(f"Model Accuracy: {accuracy * 100:.2f}%\n")
```

```python
def get_recommendations(user_genre, min_rating, user_type):
    # Start with the original filters for genre and rating
    filtered_df = df_cleaned[
        (df_cleaned['genre'].str.contains(user_genre, case=False, na=False)) &
        (df_cleaned['rating'] >= min_rating)
    ].copy()

    # If the user enters a specific type (and not 'all'), apply this filter
    if user_type.lower() != 'all':
        filtered_df = filtered_df[filtered_df['type'].str.lower() == user_type.lower()]

    if filtered_df.empty:
        return None

    # Prepare data for the model to predict on
    X_to_predict = filtered_df[['primary_genre', 'members']]
    X_to_predict_encoded = pd.get_dummies(X_to_predict, columns=['primary_genre'])
    final_cols = X_train.columns
    X_to_predict_aligned = X_to_predict_encoded.reindex(columns=final_cols, fill_value=0)

    # Make predictions and filter for the ones the model approves
    predictions = dt_model.predict(X_to_predict_aligned)
    filtered_df['prediction'] = predictions
    recommendations = filtered_df[filtered_df['prediction'] == 1].sort_values(by='rating', ascending=False)

    # Clean up sequels/duplicates from the final list
    if not recommendations.empty:
        recommendations['base_name'] = recommendations['name'].str.split(r'[:°\']').str[0].str.strip()
        recommendations.drop_duplicates(subset=['base_name'], keep='first', inplace=True)

    return recommendations[['name', 'genre', 'type', 'rating']]
```

```python
# --- INTERACTIVE RECOMMENDATION SESSION ---
print("="*50)
print("             ANIME RECOMMENDATION SYSTEM")
print("="*50 + "\n")

# --- 1. GENRE SELECTION ---
genre_list = ['Action', 'Adventure', 'Comedy', 'Drama', 'Fantasy', 'Horror', 'Mystery', 'Romance', 'Sci-Fi', 'Slice of Life', 'Sports', 'Supernatural', 'Thriller']
print("Please choose a genre from the list:")
for i, genre in enumerate(genre_list):
    print(f"  {i+1}: {genre}")
while True:
    try:
        genre_choice = int(input("Enter the number for your genre choice: "))
        if 1 <= genre_choice <= len(genre_list):
            user_favorite_genre = genre_list[genre_choice - 1]
            break
        else:
            print("Invalid number. Please choose a number from the list.")
    except ValueError:
        print("Invalid input. Please enter a number.")
print("-" * 50)

# --- 2. TYPE SELECTION ---
type_list = ['TV', 'Movie', 'OVA', 'Special', 'All']
print("Please choose the type:")
for i, type_name in enumerate(type_list):
    print(f"  {i+1}: {type_name}")
while True:
    try:
        type_choice = int(input("Enter the number for your type choice: "))
        if 1 <= type_choice <= len(type_list):
            user_type_input = type_list[type_choice - 1]
            break
        else:
            print("Invalid number. Please choose a number from the list.")
    except ValueError:
        print("Invalid input. Please enter a number.")
print("-" * 50)

# --- 3. RATING SELECTION ---
while True:
    try:
        user_min_rating = float(input("Enter the minimum rating you want (between 0.0 and 10.0): "))
        if 0.0 <= user_min_rating <= 10.0:
            break
        else:
            print("Invalid rating. Please enter a number between 0.0 and 10.0.")
    except ValueError:
        print("Invalid input. Please enter a number.")

# --- GET AND DISPLAY RECOMMENDATIONS ---
print(f"\nSearching for '{user_type_input}' anime in '{user_favorite_genre}' with a rating >= {user_min_rating}...")
print("-" * 50)

recommended_anime = get_recommendations(user_favorite_genre, user_min_rating, user_type_input)

if recommended_anime is not None and not recommended_anime.empty:
    print("Here are your top 10 recommendations:\n")
    print(recommended_anime.head(10).to_string(index=False))
else:
    print("Sorry, no recommendations found.")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# ✅ Train Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# ✅ Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# ✅ Predictions
y_pred_dt = dt_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)

# --- Decision Tree Metrics ---
mae_dt = mean_absolute_error(y_test, y_pred_dt)
mse_dt = mean_squared_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)

print("📊 Decision Tree Metrics:")
print(f"MAE  : {mae_dt:.4f}")
print(f"MSE  : {mse_dt:.4f}")
print(f"RMSE : {rmse_dt:.4f}")

# --- Random Forest Metrics ---
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)

print("\n📊 Random Forest Metrics:")
print(f"MAE  : {mae_rf:.4f}")
print(f"MSE  : {mse_rf:.4f}")
print(f"RMSE : {rmse_rf:.4f}")
```
